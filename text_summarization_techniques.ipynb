```python
## Prompt Template Text Summarization

from langchain.chains import LLMChain
from langchain import PromptTemplate

generictemplate = """
Write a summary of the following speech : 
Speech : {speech}
Translate the precise summary to {language}
"""

prompt = PromptTemplate(
    input_variables=['speech','language'],
    template = generictemplate
)

prompt
```

```python
complete_prompt = prompt.format(speech = speech , language = "hindi")
```

```python
llm.get_num_tokens(complete_prompt)
```

```python
llm_chain = LLMChain(llm=llm , prompt=prompt)
summary = llm_chain.run({'speech':speech , 'language' : 'hindi'})
```

```python
summary
```

```python
## StuffDocumentChain Text Summarization
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("/Users/sheenkocher/Desktop/generativeAI/7-Text Summarization/apjspeech.pdf")
docs = loader.load_and_split()
docs
```

```python
tempate = """
write a concise and short summary of the following speech ,
Speech : {text}
"""
prompt = PromptTemplate(input_variables=['text'],
                        template = tempate)
```

```python
from langchain.chains.summarize import load_summarize_chain
```

```python
chain = load_summarize_chain(llm,chain_type='stuff',prompt=prompt,verbose=True)
```

```python
output_summary = chain.run(docs)
output_summary = chain.run(docs)
output_summary
```

```python
print(output_summary)
```

```python
## Map reduce to summarize large documents
from langchain.text_splitter import RecursiveCharacterTextSplitter
final_docs = RecursiveCharacterTextSplitter(chunk_size = 2000 , chunk_overlap = 100 ).split_documents(docs)
```

```python
final_docs
```

```python
len(final_docs)
```

```python
prompt
```

```python
chunks_prompt = """
Please summarize the below speech :
Speech : '{text}'
Summary : 
"""

map_prompt_template = PromptTemplate(input_variables=['text'] , template = chunks_prompt)

summary_chain = load_summarize_chain(
    llm = llm ,
    chain_type= "map_reduce",
    map_prompt = map_prompt_template
)
```

```python
final_prompt = """
Provide the final summary of the entire speech with these important points .
Add a Motivational Title , Start the precise summary with an introduction and provide the summary in number
points for the speech .
Speech : {text}
"""

final_prompt_template = PromptTemplate(input_variables=['text'] , template = final_prompt)
```

```python
summary_chain = load_summarize_chain(
    llm = llm ,
    chain_type= "map_reduce",
    map_prompt = map_prompt_template,
    combine_prompt = final_prompt_template,
    verbose = True
)
```

```python
output = summary_chain.run(final_docs)
```

```python
print(output)
```

```python
### Refine Chain for Summarization 
chain = load_summarize_chain(
    llm = llm ,
    chain_type = "refine",
    verbose = True
)

output_summary = chain.run(final_docs)
```
